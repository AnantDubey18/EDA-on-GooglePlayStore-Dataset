###### Dealing with the Null Values

```python
Mean_col = ['Rating']
Mode_col = ['']
```

```python
num_col = ['Rating', 'Rating Count','Minimum Installs', 'Maximum Installs']
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
for i in df.columns:
    if i in num_col:
        sns.boxplot(df[i])
        plt.show()
```

```python
for i in df.columns:
    if i in num_col:
       df[i].fillna(df[i].median(),inplace=True)
```

```python
mode_col = []
for i in df.columns:
    if i not in num_col:
        mode_col.append(i)
print(mode_col)
```

###### FILLING REST COLUMNS WITH THE MODE

```python
for i in df.columns:
    if i in mode_col:
        df[i].fillna(df[i].mode()[0],inplace= True)
```

###### CONVERTING THE RELEASED COLUMN INTO THE DATE TIME DATA TYPE

```python
df['Released'] = pd.to_datetime(df['Released'])
```

```python
df['Released_Year'] = df['Released'].dt.year
```

```python
df['Released_Month'] = df['Released'].dt.month_name()
```

```python
df['Released_Day'] = df['Released'].dt.day_name()
```

###### DAY WISE NUMBER OF APP RELEASED
